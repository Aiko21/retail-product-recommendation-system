# Dataset Audit

---

## Objective

Validate dataset quality and identify potential issues before model development.

---

## Business Perspective

Poor data quality may produce inaccurate recommendations, resulting in poor customer experience and business losses.

Ensuring data quality is essential to build trustworthy recommendation systems.

---

## Data Science Perspective

Data auditing is the first technical step to understand dataset structure, identify anomalies, missing values, duplicates, and potential data leakage.

Poor quality data can negatively impact model performance.

---

## Methodology

The following checks will be performed:

1. Dataset loading
2. Dataset shape validation
3. Data type validation
4. Missing value assessment
5. Duplicate assessment
6. Descriptive statistics
7. Potential data leakage assessment

---

```python
# ==================================================
# IMPORT LIBRARIES
# ==================================================

from pathlib import Path
import pandas as pd
import numpy as np

pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", 100)

print("Libraries imported successfully.")
```

```python
# ==================================================
# DATASET LOCATION
# ==================================================

PROJECT_ROOT = Path().resolve().parent

DATA_PATH = PROJECT_ROOT / "data" / "raw" / "Online Retail.xlsx"

print(DATA_PATH)
print("File exists:", DATA_PATH.exists())
```

```python
# ==================================================
# LOAD DATASET
# ==================================================

df = pd.read_excel(DATA_PATH)

print("Dataset loaded successfully.")
```

```python
# ==================================================
# BASIC DATASET INFORMATION
# ==================================================

print(f"Dataset Shape: {df.shape}")

display(df.head())
```

```python
df.info()
```

```python
# ==================================================
# MISSING VALUE CHECK
# ==================================================

missing = pd.DataFrame({
    "missing_count": df.isnull().sum(),
    "missing_percentage": round(df.isnull().mean()*100,2)
})

missing = missing.sort_values(
    by="missing_percentage",
    ascending=False
)

display(missing)
```

```python
# ==================================================
# DUPLICATE CHECK
# ==================================================

duplicate_count = df.duplicated().sum()

print("Duplicate Rows:", duplicate_count)
```

```python
# ==================================================
# DATA TYPE VALIDATION
# ==================================================

dtype_df = pd.DataFrame({
    "column": df.columns,
    "dtype": df.dtypes.astype(str),
    "unique_values": [df[col].nunique() for col in df.columns]
})

display(dtype_df)
```

```python
# ==================================================
# DESCRIPTIVE STATISTICS
# ==================================================

display(df.describe(include="all").T)
```

```python
# ==================================================
# INITIAL BUSINESS OBSERVATIONS
# ==================================================

print("Total Transactions :", df["InvoiceNo"].nunique())
print("Total Customers    :", df["CustomerID"].nunique())
print("Total Products     :", df["StockCode"].nunique())
print("Countries Covered  :", df["Country"].nunique())
```

# Findings

1. The dataset contains 541,909 transaction records and 8 features.

2. Missing values were identified in:
   - CustomerID: 135,080 rows (24.93%)
   - Description: 1,454 rows (0.27%)

3. Duplicate rows were detected:
   - Total duplicate rows: 5,268

4. The dataset contains:
   - 25,900 unique transactions
   - 4,372 unique customers
   - 4,070 unique products
   - Transactions from 38 countries

5. Quantity and UnitPrice contain negative values, indicating potential product returns, cancellations, or correction transactions.

6. CustomerID is a critical feature for recommendation systems; therefore, missing CustomerID records require special attention.

7. The dataset appears to be highly dominated by transactions from the United Kingdom.

8. No immediate data leakage issue was identified at this stage, but transaction-based recommendation systems require careful handling during train-test splitting.

9. After removing cancellation transactions, no negative quantity records remained in the dataset.

10. This indicates that negative quantities were strongly associated with cancellation transactions.

---

# Decision

1. Records without CustomerID will be removed because customer-product interactions cannot be established without customer identifiers.

2. Duplicate rows will be investigated further during the data cleaning phase before deciding whether to remove them.

3. Transactions representing cancellations or returns will be analyzed separately to determine their suitability for recommendation modeling.

4. Negative quantities and negative prices will not be removed immediately; further business investigation is required.

5. Additional exploratory analysis will be conducted to understand customer behavior, transaction patterns, and country distribution.

6. Recommendation models will only be trained using valid customer-product interaction records.

7. Data cleaning decisions will be documented explicitly to ensure reproducibility.
