# Model Export for Deployment

---

## Objective

Prepare all artifacts required for Streamlit deployment.

---

## Business Perspective

Deployment-ready artifacts ensure recommendation services can be delivered to end-users consistently.

---

## Data Science Perspective

Separating training and deployment artifacts improves reproducibility, scalability, and maintainability.

---

## Methodology

1. Load trained recommendation assets.
2. Export deployment datasets.
3. Generate metadata.
4. Save deployment configuration.

---

```python
# ==================================================
# IMPORT LIBRARIES
# ==================================================

from pathlib import Path
import pandas as pd
import json
import yaml

print("Libraries imported successfully.")
```

```python
# ==================================================
# DEFINE PROJECT PATHS
# ==================================================

PROJECT_ROOT = Path().resolve().parent

DEPLOYMENT_PATH = (
    PROJECT_ROOT /
    "models" /
    "deployment"
)

DEPLOYMENT_PATH.mkdir(
    parents=True,
    exist_ok=True
)

print("Deployment folder ready.")
```

```python
# ==================================================
# LOAD INTERACTION DATASET
# ==================================================

INTERACTION_PATH = (
    PROJECT_ROOT /
    "data" /
    "features" /
    "customer_product_interactions.csv"
)

interaction_df = pd.read_csv(
    INTERACTION_PATH,
    dtype={"StockCode": str}
)

print(
    "Interaction Shape:",
    interaction_df.shape
)
```

```python
# ==================================================
# CREATE PRODUCT LOOKUP
# ==================================================

product_lookup = (

    interaction_df

    .groupby("StockCode")["Description"]

    .first()

    .reset_index()

)

display(
    product_lookup.head()
)

print(
    "Total Products:",
    product_lookup.shape[0]
)
```

```python
# ==================================================
# SAVE PRODUCT LOOKUP
# ==================================================

product_lookup.to_csv(

    DEPLOYMENT_PATH /
    "product_lookup.csv",

    index=False

)

print(
    "product_lookup.csv saved successfully."
)
```

```python
# ==================================================
# LOAD BASELINE RECOMMENDATIONS
# ==================================================

BASELINE_PATH = (
    PROJECT_ROOT /
    "models" /
    "baseline" /
    "popularity_recommendations.csv"
)

baseline_df = pd.read_csv(
    BASELINE_PATH,
    dtype={"StockCode": str}
)

display(
    baseline_df.head()
)
```

```python
# ==================================================
# SAVE POPULAR PRODUCTS
# ==================================================

baseline_df.to_csv(

    DEPLOYMENT_PATH /
    "top_popular_products.csv",

    index=False

)

print(
    "top_popular_products.csv saved successfully."
)
```

```python
# ==================================================
# GENERATE METADATA
# ==================================================

metadata = {

    "project_name":
        "Retail Product Recommendation System",

    "project_type":
        "Item-Based Collaborative Filtering",

    "dataset":
        "Online Retail Dataset",

    "total_products":
        int(product_lookup.shape[0]),

    "recommendation_model":
        "Cosine Similarity",

    "baseline_model":
        "Popularity-Based Recommender",

    "evaluation_metric":

        {

            "best_k": 5,

            "precision_at_k": 0.019289,

            "recall_at_k": 0.014063,

            "map_at_k": 0.010651

        }

}
```

```python
# ==================================================
# SAVE METADATA
# ==================================================

with open(

    DEPLOYMENT_PATH /
    "metadata.json",

    "w"

) as f:

    json.dump(

        metadata,

        f,

        indent=4

    )

print(
    "metadata.json saved successfully."
)
```

```python
# ==================================================
# CREATE DEPLOYMENT CONFIG
# ==================================================

deployment_config = {

    "app_name":
        "Retail Recommendation App",

    "recommendation_top_n":
        5,

    "cold_start_strategy":
        "popularity_based",

    "similarity_threshold":
        0.60

}
```

```python
# ==================================================
# SAVE DEPLOYMENT CONFIG
# ==================================================

with open(

    DEPLOYMENT_PATH /
    "deployment_config.yaml",

    "w"

) as f:

    yaml.dump(

        deployment_config,

        f,

        sort_keys=False

    )

print(
    "deployment_config.yaml saved successfully."
)
```

```python
# ==================================================
# VALIDATE DEPLOYMENT FILES
# ==================================================

for file in DEPLOYMENT_PATH.iterdir():

    print(file.name)
```

# Findings

- Deployment artifacts were successfully exported.
- Product lookup tables were prepared for the web application.
- Baseline recommendations were preserved to address cold-start scenarios.

---

# Decision

- The exported artifacts will be used by the Streamlit application.
- Popularity-based recommendations will be used as fallback recommendations.

---

# Business Insight

Well-structured deployment artifacts simplify production deployment and improve maintainability.
